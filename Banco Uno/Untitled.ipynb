```python
plt.scatter(credit["Credito"],credit["MFSEP05"])
plt.xlabel("Credito")
plt.ylabel("MFSEP05")
plt.show()
```

```python
msk = np.random.rand(len(cdf1)) < 0.8 #Esto selecciona aleatoriamente el 80% de las observaciones y la amacena en msk
train = cdf1[msk] #aca se indica que el set de entrenamiento esta conformado por el 80% de las observaciones
test = cdf1[~msk]
```

```python
plt.scatter(train.Credito, train.MFSEP05, color='blue')
plt.xlabel("Credito")
plt.ylabel("MFSEP05")
plt.show()
```

```python
cdf1= linear_model.LinearRegression()
train_x = np.asanyarray(train[['Credito']])
train_y = np.asanyarray(train[["MFSEP05"]])
cdf1.fit(train_x, train_y)
# Obtener coeficientes de resultados, estos datos son informativos.
print ('Coefficients: ', cdf1.coef_)
print ('Intercept: ',cdf1.intercept_)
```

```python
plt.scatter(train.Credito, train.MFSEP05, color='blue')
plt.plot(train_x, cdf1.coef_[0][0]*train_x + cdf1.intercept_[0], '-r')
plt.xlabel("Credito")
plt.ylabel("MFSEP05")
```

```python
from sklearn.metrics import r2_score
test_x = np.asanyarray(test[['Credito']])
test_y = np.asanyarray(test[["MFSEP05"]])
test_y_ = cdf1.predict(test_x)
print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_ - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_ - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y , test_y_))
```

```python
cdf1 = pd.DataFrame({'Actual': test_y.flatten(), 'Predicted': test_y_.flatten()})
cdf1.head(10)
```
